```python
import torch
from IPython.display import Image, display

# Load the trained model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/yolov5/runs/train/bottle_detection2/weights/best.pt')

# Perform inference on an image
img = '/content/dataset/valid/images/00000000_jpg.rf.01d5997f53766d45837c6c375df9aaac.jpg'  # Replace with the path to your image
results = model(img)

# Print and inspect the results
print("Detected objects:")
results.print()

# Iterate through results and print detailed information
for i, det in enumerate(results.xyxy[0]):
    xmin, ymin, xmax, ymax, conf, cls = det
    print(f"Detection {i + 1}:")
    print(f"Class: {model.names[int(cls)]}")
    print(f"Confidence: {conf:.2f}")
    print(f"Bounding Box: {xmin:.2f}, {ymin:.2f}, {xmax:.2f}, {ymax:.2f}\n")

# Save the results image
results.save()

# Display the output image using the correct path
output_image_path = 'runs/detect/exp/00000000_jpg.rf.01d5997f53766d45837c6c375df9aaac.jpg'
display(Image(filename=output_image_path))
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread(img)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Plot the bounding boxes and labels
for *box, conf, cls in results.xyxy[0]:  # xyxy format
    label = f"{model.names[int(cls)]} {conf:.2f}"
    cv2.rectangle(image, (int(box[0]), int(box[1])), (int(box[2]), int(box[3])), (255, 0, 0), 2)
    cv2.putText(image, label, (int(box[0]), int(box[1])-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

# Show the image
plt.imshow(image)
plt.axis('off')
plt.show()
```
